1. Lognormal vs. Normal Distribution

Below is a code snippet of a well-fitting lognormal (reversed) distribution to the visibility function, $v(z)$. This is indeed a good idea as the lognormal distribution is skewed to one side and this matches the theoretical prediction of the actual visibility function. Here are a few reasons why we think it is better to use a Gaussian:

   a) it is easier to interpret physically (i.e. the $\sigma$ which appears in the Gaussian form has an actual physical interpretation as the duration of last scattering) 
    
   b) it does not make a huge difference, since in the approximation where the parameter $\sigma$ is small, the two distributions are very similar (see below) and we are only constraining one parameter (effectively, the second moment). If working with conformal time, $\eta$, for which the scaling is different, the Gaussian fit is even more similar to the lognormal one.
    
   c) it does not assume the predicted asymmetry, allowing for a more general probe of the function's shape.

```python
from scipy.stats import lognorm
import matplotlib.pyplot as plt
import numpy as np
```

```python
# parameters for the lognormal
s = 0.1
mu = 7.
scale = np.exp(mu)

# parameters for the Gaussian
m = 1086.73
sig = 120

# which values to plot over
z = np.linspace(0.01,2*m,500)

# find max of lognormal
fig, ax = plt.subplots(1, 1,figsize = (11,7))
lognor = lognorm.pdf(z, s,scale=scale)
max = np.max(lognor)
lognor /= max
argmax = np.argmax(lognor)
print(z[argmax])

# normal or gaussian distn
gauss = 1/(sig * np.sqrt(2 * np.pi))*np.exp( - (z - m)**2 / (2 * sig**2))
gauss /= np.max(gauss)

# plotting both (notice that lognor is backwards because we want it to be skewed to the left)
ax.plot(z, gauss,
        'b-', lw=2.5, alpha=0.6, label='norm pdf')
ax.plot(z, (lognor[::-1]),
        'r-', lw=2.5, alpha=0.6, label='lognorm pdf')
plt.xlim([600,1600])
plt.xlabel(r"$z$")
plt.ylabel(r"$v(z)$")
ax.legend()
```

3. Quantitative prediction for Stage-4

Thank you for pointing this out! We can make a quantitative prediction about Stage-4 experiments in the following way:

Let us take our more stringent constraint, which comes from using both the temperature CMB measurements and the polarization measurements.

We model the noise of the polarization measurement of Planck as $n = 50 \mu$K-arcmin and $\theta_{FWHM} = 5$' up to $\ell = 2000$
and the noise of CMB-S4 as $n = \sqrt 2 \mu$K-arcmin and $\theta_{FWHM} = 1$' up to $\ell = 4000$.

We then compute S/N using the following code:

```python
import numpy as np
import matplotlib.pyplot as plt

#noiseFactor = 1.
noiseFactor = 50.

if noiseFactor == 1.: # CMB-S4 conservative
    thetaFWHMarcmin = 1.
    noiseFactor *= np.sqrt(2.)    
    L = 4000
    
if noiseFactor == 50.: # Planck 
    thetaFWHMarcmin = 5.
    L = 2000
    
# just aids conversion
noiseUkArcmin = 1.                      
thetaFWHM = thetaFWHMarcmin*np.pi/(180.*60.) 
deltaT = noiseUkArcmin/thetaFWHMarcmin
f_sky = .5

fname = "/home/boryanah/PRD/SNR/Aug6_highAcc_CDM_scalCls.dat"
CL = np.loadtxt(fname)
ell = CL[:,0]
CEE = CL[:,2]

N_ellP = (deltaT*thetaFWHM)**2*np.exp(ell*(ell+1.)*thetaFWHM**2/(8.*np.log(2.)))*noiseFactor**2.
C_ellE = CEE/(ell*(ell+1.)/(2*np.pi))
DC_ellE = np.sqrt(2./(2*ell+1)/f_sky)*(C_ellE + N_ellP)

print(ell[L])
snr = C_ellE/DC_ellE
snr = np.sqrt(np.sum(snr[1:L]**2))
print(snr)

plt.plot(ell,DC_ellE*ell*(ell+1)/(2*np.pi))
plt.plot(ell,CEE)
plt.xlim([0,4000])
plt.ylim([0,50])
```

We find that the signal-to-noise for the two experiments is 1550 and 302, so the expected improvement is a 5 times smaller error or 19$\pm $0.013 Mpc.

4. Testing the functional form of the visibility function

We can generalize our Gaussian assumption by considering a fit of the form:
\begin{equation}
\frac{\beta}{2\alpha\Gamma(1/\beta)} \; e^{-(|x-\mu|/\alpha)^\beta}.
\end{equation}
One can then show that the mean of this distribution is $\mu$ for all $\alpha$
and $\beta$, and the variance is:
\begin{equation}
{\frac {\alpha ^{2}\Gamma (3/\beta )}{\Gamma (1/\beta )}},
\end{equation}
so this family of generalized normal distributions share the same property as the Gaussian, as the variance is proportional to the standard width $\sigma$.

Since in this paper, we are only interested in constraining the second moment of the visibility function (in order to shed light on a new method of doing so through the polarization-to-temperature ratio), it is sufficient to add a single parameter. Exploring more exotic shapes of the visibility function requires more parameters and is potentially left for future work.
